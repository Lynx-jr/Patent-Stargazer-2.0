## Network analysis

```python
import pandas as pd 

df = pd.read_csv('/content/drive/MyDrive/finalcount.csv')
df_ref = pd.read_csv('/content/drive/MyDrive/cpc_subsection.tsv', sep='\t')
```

```python
import networkx as nx
G = nx.from_pandas_edgelist(df[df['count']>100000], # 100,000
                            source = 'subsection_id',
                            target = 'citation_subsection_id',
                            edge_key = 'count',
                            create_using=nx.DiGraph())
```

```python
import matplotlib.pyplot as plt
import networkx as nx
pos = nx.circular_layout(G)
# # default
# plt.figure(1)
# nx.draw_networkx(G,pos)
# # smaller nodes and fonts
# plt.figure(2)
# nx.draw_networkx(G,pos,node_size=60,font_size=8) 
# # larger figure size
plt.figure(3,figsize=(12,12)) 
nx.draw_networkx(G,pos)
plt.show()
```
